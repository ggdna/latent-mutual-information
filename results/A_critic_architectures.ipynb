```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
from scipy.stats import kendalltau
import numpy as np

rcParams['figure.dpi'] = 150
```

```python
df = pd.read_csv('A_critic.csv', index_col=0)
df['Dims.'] = df['Dimensions']
df['Estimator'] = df['Measure']
df['Estimate'] = [max(0, x) for x in df['Estimate'].values]
```

```python
sns.set_style('whitegrid')
rcParams['figure.dpi'] = 150
rcParams['figure.figsize'] = (4, 3)
rcParams['grid.linewidth'] = 0.5
rcParams['grid.linestyle'] = 'dashed'
rcParams['legend.fancybox'] = False
rcParams['mathtext.fontset'] = 'stix'
palette=['#5A5B9F', '#45B5AA']

sns.relplot(df, x='Max layer size', y='Estimate', 
            hue='Estimator', col='Dims.', kind='line', aspect=1, height=2, lw=3,
           palette=palette)
plt.ylim(0, 1)
```
